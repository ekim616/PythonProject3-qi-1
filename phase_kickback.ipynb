```
### Phase kickback
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_state_qsphere  # added; EK

s = '10'
num_qubits = len(s)
circ = QuantumCircuit(num_qubits)  # added; EK
# Set reverse ordering
s = s[::-1]
# Construct the state 110
for idx in range(num_qubits):
    if s[idx] == '1':
        circ.x(idx)
circ.barrier()
circ.draw(output='mpl')

display(circ.draw('latex'))
state = Statevector(circ)
display(plot_bloch_multivector(state, reverse_bits=True))
display(plot_state_qsphere(state))  # added; EK

# Added version; EK
display(state.draw(output='latex'))  # added; EK

#circ = QuantumCircuit(2)
circ.x(1)
circ.h(0)
circ.h(1)
display(circ.draw('latex'))
circ.draw(output='mpl')  # added; EK

state = Statevector(circ)
display(plot_bloch_multivector(state, reverse_bits=True))
display(plot_state_qsphere(state))  # added; EK

# Added version; EK
display(state.draw(output='latex'))
```
